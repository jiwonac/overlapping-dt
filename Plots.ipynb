```python
# Import libraries for plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Results of experiment 1
x = [5, 10, 20, 30, 40, 50]
classic = [8.5, 16.1, 44.7, 77.3, 138.8, 352.4]
ours = [7.6, 18.0, 47.6, 89.7, 133.4, 190.7]
random = [16.8, 33.6, 87.6, 160.2, 288.2, 776.0]
```

```python
plt.plot(x, classic, label = "classic")
plt.plot(x, ours, label = "ours")
plt.plot(x, random, label = "random")
plt.legend()
plt.xticks(x)
plt.xlabel("group requirements")
plt.ylabel("average cost required to obtain one tuple")
plt.title("Default parameters, 20% overlap, all binary.")
plt.show()
```
